```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lines=[]
with open('bipolar_momentum_XOR.csv',"r") as inp:
    lines=[line.split(",") for line in inp]
```

```python
data = np.asarray(lines)
```

```python
epochs=[]
for i in data:
    epochs.append(int(i[0]))
```

```python
epochs
```

```python
epochs=np.array(epochs)
print("Fastest Convergence ",np.amin(epochs))
print("Slowest Convergence ",np.amax(epochs))
print("Average Convergence ",np.average(epochs))
```

```python
graph=[]
for i  in range(len(data[10])-1):
    graph.append(float(data[10][i]))
print(graph[0])
```

```python
graph_slow=[]
graph_fast=[]
graph_avg=[]
for i  in range(len(data[73])-1):
    graph_slow.append(float(data[73][i]))
for i in range(len(data[1])-1):
    graph_fast.append(float(data[1][i]))
for i in range(len(data[10])-1):
    graph_avg.append(float(data[10][i]))
```

```python
x_avg=[]
x_fast=[]
x_slow=[]
for i in range(len(graph_avg)-1):
    x_avg.append(i)
for i in range(len(graph_fast)-1):
    x_fast.append(i);
for i in range(len(graph_slow)-1):
    x_slow.append(i)
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(10,6))
plt.title("Total Error v/s Number of Epochs")
plt.plot(x_avg,graph_avg[1:],c='b',label="Average(Bipolar, momentum=0.9)")
plt.plot(x_slow,graph_slow[1:],c='r',label="Slow(Bipolar, momentum=0.9)")
plt.plot(x_fast,graph_fast[1:],c='g',label="Fast(Bipolar),momentum=0.9)")
plt.xlabel("Number of Epochs",fontsize=18)
plt.ylabel("Total Error",fontsize=18)
plt.legend(loc="best")
plt.savefig("Bipolar_momentum.png")
```
